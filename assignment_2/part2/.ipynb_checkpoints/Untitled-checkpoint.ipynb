```python
import matplotlib.pyplot as plt

s = 100
r = 0.25
a = 0
tau = 250
k = []
for t in range(500):
    phi = ((t-s)% tau) / tau
    
    if phi < (0.5*r):
        k.append(2*phi/r)
    elif phi > (0.5*r) and phi < r:
        k.append(2 - (2*phi/r))
    else:
        k.append(a*phi)



plt.plot(k)        
arrow1 = plt.arrow(0,1.4,100,0, color='red', label ='s', width=0.01, head_length=.5, head_width=.2)
arrow2 = plt.arrow(100,1.7,250,0, color='green', label = 'tau', width=0.01, head_length=.5, head_width=.2)
arrow3 = plt.arrow(100,1.1,65,0, color='orange', label = 'rho', width=0.01, head_length=.5, head_width=.2)
plt.arrow(100,1.4,-100,0, color='red', label ='s', width=0.01, head_length=.5, head_width=.2)
plt.arrow(250,1.7,-150,0, color='green', label = 'tau', width=0.01, head_length=.5, head_width=.2)
plt.arrow(125,1.1,-25,0, color='orange', label = 'rho', width=0.01, head_length=.5, head_width=.2)
plt.ylim(-0.1,2)
plt.xlabel('t')
plt.ylabel('k')

plt.legend([arrow1,arrow2,arrow3], ['s','tau','$r_{on}$'])
plt.show()

        
    

```
